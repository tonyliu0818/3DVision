```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

img1 = plt.imread('box_left_2k.jpg')
img2 = plt.imread('box_right_2k.jpg')

pts1 = np.array([[1148, 351], [1441, 481], [1427, 662], [1172, 525], [1524,80], [1795,134], [1744,307],
        [1564, 320], [1579, 306], [1584, 288]])

pts2 = np.array([[747, 397], [944,522], [1010,704], [822,558], [1212,116], [1440,170], [1433,350],
        [1131,360], [1152,342], [1166,326]])

K1 = np.array([[1062.04, 0, 1098.52], [0, 1061.74, 599.665], [0, 0, 1]])
K2 = np.array([[1060.59, 0, 1127.03], [0, 1060.06, 645.711], [0, 0, 1]])

# TODO 1 (15 points): Normalize image points.
#        There is no need to use for loops in this task. If you use for loops in this task, you can get at most 10 point.
print("TODO 1:")
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.scatter(pts1[:, 0], pts1[:, 1], color='red', marker='o')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('pts1')

#p_ = mean(pts) = tx,ty
pts2_mean = np.mean(pts2, axis=0)
pts1_mean = np.mean(pts1, axis=0)

ax1.scatter(pts1_mean[0],pts1_mean[1], color='blue', marker='o')

#qi = pi  p_.
pts1_normal = pts1 - pts1_mean 
pts2_normal = pts2 - pts2_mean

mean_distance1 = np.mean(np.sqrt(np.sum(pts1_normal ** 2, axis=1)))
mean_distance2 = np.mean(np.sqrt(np.sum(pts2_normal ** 2, axis=1)))
s1 = np.sqrt(2) / mean_distance1
s2 = np.sqrt(2) / mean_distance2

#pts_normal = s*qi
pts1_normal = s1 * pts1_normal
pts2_normal = s2 * pts2_normal
ax2.scatter(pts1_normal[:, 0], pts1_normal[:, 1], color='red', marker='o')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('pts1_normal')
pts1_normal_mean = np.mean(pts1_normal, axis=0)
ax2.scatter(pts1_normal_mean[0],pts1_normal_mean[1], color='blue', marker='o')
plt.show()

tx,ty = pts1_mean
tx_,ty_ = pts2_mean
T1 = np.array([[s1,  0, -s1 *  tx],
               [ 0, s1, -s1 *  ty],
               [ 0,  0,         1]])
T2 = np.array([[s2,  0, -s2 * tx_],
               [ 0, s2, -s2 * ty_],
               [ 0,  0,         1]])
mean_distance_todo1_6 = np.mean(np.sqrt(np.sum(pts1_normal ** 2, axis=1)))
print(f"The mean distance from pts1_normal to the origin: {mean_distance_todo1_6}")
# TODO 2 (25 points): Compute the fundamental matrix F.
print("\nTODO 2:")
A=[]
for i in range(pts1_normal.shape[0]):
    x,y = pts1_normal[i]
    x_,y_ = pts2_normal[i]
    A.append([x*x_, x_*y, x_, x*y_, y*y_, y_, x, y, 1]) 
A=np.array(A,dtype=np.float64)
U, S, VT = np.linalg.svd(A)
f = VT.T[:,-1]
F = f.reshape(3,3)
#print(f"F rank: {np.linalg.matrix_rank(F)}") 3
U2, S2, VT2 = np.linalg.svd(F)
F = U2 @ np.diag([S2[0], S2[1], 0]) @ VT2
print(f"F rank: {np.linalg.matrix_rank(F)}")

F = T2.T @ F @ T1

epipolar_error = 0
for i in range(pts1.shape[0]):
    point1 = np.append(pts1[i], 1)#[xi, yi, 1]
    point2 = np.append(pts2[i], 1) 
    epipolar_error += np.abs(point2 @ F @ point1.T)
print("Epipolar Error:", epipolar_error)
if epipolar_error < 1:
    print("Epipolar error is less than one. Nice!!")
else:
    print("there is something wrong with your program.")
# TODO 3 (15 points): Draw epipolar lines in image1.
print("\nTODO 3:")
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

p1_ = pts2[0]
p2_ = pts2[1]
p3_ = pts2[2]

p_list = [p1_, p2_, p3_]
colors = ['steelblue', 'darkorange', 'green']

ax1.imshow(img1)
ax2.imshow(img2)
for i, p_ in enumerate(p_list):
    line = F.T @ np.append(p_, 1)
    a, b, c = line

    x = np.linspace(0, img1.shape[1]-1, 3)
    y = (-a * x - c) / b
    
    ax1.plot(x, y, color=colors[i], linewidth=2)    
    ax2.scatter(p_[0], p_[1], color=colors[i], marker='o', s=10)

ax1.set_title('left image')
ax2.set_title('right image')
ax1.axis('off')
ax2.axis('off')
plt.show()
# TODO 4 (10 points): Draw the epipole and epipolar lines in image1.
print("\nTODO 4:")
fig, ax = plt.subplots(figsize=(12, 6))

U, S, VT = np.linalg.svd(F)
epipole = VT[-1]
epipole /= epipole[2]
ax.imshow(img1)

for i, p_ in enumerate(p_list):
    line = F.T @ np.append(p_, 1)
    a, b, c = line
    
    x = np.linspace(0, epipole[0], 3)
    y = (-a * x - c) / b
    
    ax.plot(x, y, color=colors[i], linewidth=3)

ax.scatter(epipole[0], epipole[1], color='red', marker='x', s=100)
ax.axis('off')
ax.set_title('epipole and epipolar lines in image1')
plt.show()
# TODO 5 (5 points): Compute the 4 possible camera matrices.
#print("\nTODO 5:")
E = K2.T @ F @ K1
U, S, VT = np.linalg.svd(E)
u3 = U[:,-1]
W = np.array([[0,-1, 0],
              [1, 0, 0],
              [0, 0, 1]])
#print(f"W = {np.linalg.det(U @ W @ VT)}")
if(np.linalg.det(U @ W @ VT) < 0):
    W *= -1
R1 = U @ W @ VT
R2 = U @ W.T @ VT
t1 = u3
t2 = -u3

Rt1 = np.hstack((R1, t1.reshape(3, 1)))
Rt2 = np.hstack((R1, t2.reshape(3, 1)))
Rt3 = np.hstack((R2, t1.reshape(3, 1)))
Rt4 = np.hstack((R2, t2.reshape(3, 1)))
Rt2_list = [Rt1, Rt2, Rt3, Rt4]

# TODO 6 (25 points): Triangulation.
#        There is no need to use for loops when computing the reconstruction errors of pts1 (or pts2).
#        If you do, you can get at most 20 point.
print("\nTODO 6:")
X_list = []
for j in range(4):
    P1 = K1 @ np.eye(3,4)
    P2 = K2 @ Rt2_list[j]
    X = np.zeros((10, 3))  # 
    for i in range(10):
        xi, yi = pts1[i]
        xi_, yi_ = pts2[i]
        A = np.array([[P1[0, 0] - P1[2, 0] * xi , P1[0, 1] - P1[2, 1] * xi , P1[0, 2] - P1[2, 2] * xi ],
                      [P1[1, 0] - P1[2, 0] * yi , P1[1, 1] - P1[2, 1] * yi , P1[1, 2] - P1[2, 2] * yi ],
                      [P2[0, 0] - P2[2, 0] * xi_, P2[0, 1] - P2[2, 1] * xi_, P2[0, 2] - P2[2, 2] * xi_],
                      [P2[1, 0] - P2[2, 0] * yi_, P2[1, 1] - P2[2, 1] * yi_, P2[1, 2] - P2[2, 2] * yi_]])

        b = np.array([[P1[2, 3] * xi  - P1[0, 3]],
                      [P1[2, 3] * yi  - P1[1, 3]],
                      [P2[2, 3] * xi_ - P2[0, 3]],
                      [P2[2, 3] * yi_ - P2[1, 3]]])
        Mi = np.linalg.lstsq(A, b, rcond=None)[0]
        X[i] = Mi.flatten()  
    X_list.append(X)

positive_j = None
X_best = None

for j, X in enumerate(X_list):
    if np.all(X[:, 2] > 0):
        positive_j = j
        X_best = X
        break
        
P1 = K1 @ np.eye(3,4)
P2 = Rt2_list[positive_j]
Mi = X_best
Mi_ = np.ones((Mi.shape[0], 4))
Mi_[:, :3] = Mi
m_1 = P1 @ Mi_.T
m_2 = P2 @ Mi_.T
m1 = np.array([m_1[0,:]/m_1[2,:],m_1[1,:]/m_1[2,:]]).T
m2 = np.array([m_2[0,:]/m_2[2,:],m_2[1,:]/m_2[2,:]]).T

reconstruction_error_pts1 = np.mean(np.sqrt((pts1[0]-m1[0])**2 + (pts1[1]-m1[1])**2))
reconstruction_error_pts2 = np.mean(np.sqrt((pts2[0]-m2[0])**2 + (pts2[1]-m2[1])**2))
print(f"The reconstruction error for pts1: {reconstruction_error_pts1}")
print(f"The reconstruction error for pts2: {reconstruction_error_pts2}")
# TODO 7 (5 points): Visualize the 3D reconstruction.
print("\nTODO 7:")
fig = plt.figure(figsize=(4, 4))
fig4_p = pts1.copy()
fig4_p[:, 0] -= 1000
for i, pt in enumerate(fig4_p):
    plt.text(pt[0], pt[1],f'{i}')
plt.imshow(img1[:800, 1000:2000])
plt.axis('off')
plt.show()

edges = [(0, 1), (1, 2), (2, 3), (3, 0), (4, 5), (5, 6), (0, 4), (1, 5), (2, 6)]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6), subplot_kw={'projection': '3d'})

#ax.view_init(elev=-90, azim=-115)
ax1.scatter(X_best[:, 0], X_best[:, 1], X_best[:, 2], c='b', marker='o')

for edge in edges:
    ax1.plot([X_best[edge[0], 0], X_best[edge[1], 0]],
             [X_best[edge[0], 1], X_best[edge[1], 1]],
             [X_best[edge[0], 2], X_best[edge[1], 2]], 'black')

ax1.set_title('View 1: elev=-90, azim=-115')
ax1.view_init(elev=-90, azim=-115)

#ax.view_init(elev=-139, azim=-113)
ax2.scatter(X_best[:, 0], X_best[:, 1], X_best[:, 2], c='b', marker='o')

for edge in edges:
    ax2.plot([X_best[edge[0], 0], X_best[edge[1], 0]],
             [X_best[edge[0], 1], X_best[edge[1], 1]],
             [X_best[edge[0], 2], X_best[edge[1], 2]], 'black')

ax2.set_title('View 2: elev=-139, azim=-113')
ax2.view_init(elev=-139, azim=-113)
plt.show()
```
