```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math

img1 = plt.imread('box_left_2k.jpg')
img2 = plt.imread('box_right_2k.jpg')

#每一個座標點
pts1 = np.array([[1148, 351], [1441, 481], [1427, 662], [1172, 525], [1524,80], [1795,134], [1744,307],
        [1564, 320], [1579, 306], [1584, 288]])

pts2 = np.array([[747, 397], [944,522], [1010,704], [822,558], [1212,116], [1440,170], [1433,350],
        [1131,360], [1152,342], [1166,326]])
#相機參數
K1 = np.array([[1062.04, 0, 1098.52], [0, 1061.74, 599.665], [0, 0, 1]])
K2 = np.array([[1060.59, 0, 1127.03], [0, 1060.06, 645.711], [0, 0, 1]])

# TODO 1 (15 points): Normalize image points.
#        There is no need to use for loops in this task. If you use for loops in this task, you can get at most 10 point.
#把pts1取平均值
m1=np.mean(pts1,axis=0)
#q1=pts1-平均值
q1=(pts1-m1)
#s=sqrt(2)*n/q的範數
s1=math.sqrt(2)/np.mean(np.sqrt(np.sum((q1)**2,axis=1)))
#算t1
t1=np.array([[s1,0,-s1*m1[0]],[0,s1,-s1*m1[1]],[0,0,1]])
#pts1_norm=s*q
pts1_normal=s1*q1
#mean_distance
mean_distance1=np.linalg.norm(pts1_normal,axis=1).mean()
print(f'mean_distance1:{mean_distance1}')
m2=np.mean(pts2,axis=0)
q2=(pts2-m2)
s2=math.sqrt(2)/np.mean(np.sqrt(np.sum((q2)**2,axis=1)))
t2=np.array([[s2,0,-s2*m2[0]],[0,s2,-s2*m2[1]],[0,0,1]])
pts2_normal=s2*q2
mean_distance2=np.linalg.norm(pts2_normal,axis=1).mean()
print(f'mean_distance2:{mean_distance2}')
```

```python
# TODO 2 (25 points): Compute the fundamental matrix F.
def calc_fundamental(pts1,pts2):
  #創建一個array A
  A=[]
  for i in range(pts1.shape[0]):
    #第一張圖的座標點
    x1,y1=pts1[i]
    #第二張圖的座標點
    x2,y2=pts2[i]
    #代入投影片的公式
    A.append([x1*x2,x2*y1,x2,x1*y2,y1*y2,y2,x1,y1,1])
  #對a取SVD，並將v轉至
  u1,d1,v1t=np.linalg.svd(A)
  v1=np.transpose(v1t)
  #取出最後一個column並把他reshape成3*3
  f=v1[:,-1].reshape(3,3)
  #再對f做一次svd，因為出來的矩陣要rank=2
  u2,d2,v2t=np.linalg.svd(f)
  F=u2@np.array([[d2[0],0,0],[0,d2[1],0],[0,0,0]])@v2t
  return F
Fn=calc_fundamental(pts1_normal,pts2_normal)
#todo2的第3點
F=np.transpose(t2)@Fn@t1
#根據todo2的第4點，算出error
sum=0
for i in range(pts1.shape[0]):
  x1,y1=pts1[i]
  x2,y2=pts2[i]
  sum+=np.array([x2,y2,1])@F@np.array([x1,y1,1])
print(f'epipolar error:{sum}')
```

```python
# TODO 3 (15 points): Draw epipolar lines in image1.
fig=plt.figure()
#右邊的那一張圖
plt.subplot(1,2,2)
plot_img2=img2.copy()
#畫出p1'p2'p3'
plt.scatter(pts2[0,0],pts2[0,1],c='b',s=2)
plt.scatter(pts2[1,0],pts2[1,1],c='orange',s=2)
plt.scatter(pts2[2,0],pts2[2,1],c='g',s=2)
plt.xticks([])
plt.yticks([])
plt.xlabel('Figure 2: right image')
plt.imshow(plot_img2)
#左邊的那一張圖
plt.subplot(1,2,1)
#代入todo的hint第一點的公式
line1=np.transpose(F)@np.array([pts2[0,0],pts2[0,1],1])
#fig1裡的x，為了要在圖片中畫線
x1=np.arange(img1.shape[1])
#ax+by+c=0 -> y=(-c-ax)/b ->把x代進式子裡就可以得到y座標，line[0]=a，line[1]=b，line[2]=c
y1=(-line1[0]*x1-line1[2])/line1[1]
line_img1=img1.copy()
#在左邊的圖畫出藍線
plt.plot(x1,y1,c='b')
line2=np.transpose(F)@np.array([pts2[1,0],pts2[1,1],1])
x2=np.arange(img1.shape[1])
y2=(-line2[0]*x2-line2[2])/line2[1]
#在左邊的圖畫出橘線
plt.plot(x2,y2,c='orange')
line3=np.transpose(F)@np.array([pts2[2,0],pts2[2,1],1])
x3=np.arange(img1.shape[1])
y3=(-line3[0]*x3-line3[2])/line3[1]
#在左邊的圖畫出綠線
plt.plot(x3,y3,c='g')
plt.xticks([])
plt.yticks([])
plt.xlabel('Figure 1: left image')
plt.imshow(line_img1)
plt.show()
```

```python
# TODO 4 (10 points): Draw the epipole and epipolar lines in image1.
#因Fe=0，所以e為F的null space，所以要對F做svd
u,d,vt=np.linalg.svd(F)
v=np.transpose(vt)
#最後一個column為epipole
epipole=v[:,-1]
#原本是[a,b,c]->[a/c,b/c]
e=np.array([int(epipole[0]/epipole[2]),int(epipole[1]/epipole[2])])
#叫出一張畫布，裡面全是白色的。e[0]代表epipole所在的x軸，加100為了美觀。
white_img=np.ones(shape=(img1.shape[0],e[0]+100,img1.shape[2]),dtype=np.uint8)*255
#把左邊填滿img1
white_img[:,:img1.shape[1]]=img1
fig=plt.figure()
#大張畫布裡面的x
x4=np.arange(e[0])
#依據點斜式代入
y4=(-line1[0]*x4-line1[2])/line1[1]
#劃出藍線
plt.plot(x4,y4,c='b')
x5=np.arange(e[0])
#劃出橘線
y5=(-line2[0]*x5-line2[2])/line2[1]
plt.plot(x5,y5,c='orange')
##劃出綠線
x6=np.arange(e[0])
y6=(-line3[0]*x6-line3[2])/line3[1]
plt.plot(x6,y6,c='g')
#把epipole的位置畫出來
plt.scatter(e[0],e[1],c='r',marker='x')
plt.xticks([])
plt.yticks([])
plt.xlabel('Figure 3')
plt.imshow(white_img)
plt.show()
```

```python
# TODO 5 (5 points): Compute the 4 possible camera matrices.
#以下為page22的內容
E=np.transpose(K2)@F@K1
#對e做svd
u,d,vt=np.linalg.svd(E)
#u3為u的最後一個column
u3=u[:,-1]
w=np.array([[0,-1,0],[1,0,0],[0,0,1]])
#det<0的話w=-w
if np.linalg.det(u@w@vt)<0:
  w=-w
#p1=k1[I|0]  
p1=K1@np.hstack((np.identity(3),np.zeros(shape=(3,1))))
#定義r1 r2 t1 t2
r1=u@w@vt
r2=u@np.transpose(w)@vt
t1=np.expand_dims(u3,axis=1)
t2=np.expand_dims(-u3,axis=1)
#把4個可能的結果變成一個list
Rt2_list=[]
Rt2_list.append(np.hstack((r1,t1)))
Rt2_list.append(np.hstack((r1,t2)))
Rt2_list.append(np.hstack((r2,t1)))
Rt2_list.append(np.hstack((r2,t2)))
```

```python
# TODO 6 (25 points): Triangulation.
#        There is no need to use for loops when computing the reconstruction errors of pts1 (or pts2).
#        If you do, you can get at most 20 point.
#跑4個可能結果看哪一個正確
for i in range(4):
  #p2=k2[R|t]，[R|t] is from Rt2_list
  p2=K2@Rt2_list[i]
  X=[]
  for j in range(pts1.shape[0]):
    #第1和第2張圖的座標
    x1,y1=pts1[j]
    x2,y2=pts2[j]
    #根據投影片page7製造矩陣
    Ai=np.array([[p1[0,0]-p1[2,0]*x1,p1[0,1]-p1[2,1]*x1,p1[0,2]-p1[2,2]*x1],
        [p1[1,0]-p1[2,0]*y1,p1[1,1]-p1[2,1]*y1,p1[1,2]-p1[2,2]*y1],
        [p2[0,0]-p2[2,0]*x2,p2[0,1]-p2[2,1]*x2,p2[0,2]-p2[2,2]*x2],
        [p2[1,0]-p2[2,0]*y2,p2[1,1]-p2[2,1]*y2,p2[1,2]-p2[2,2]*y2]
    ])
    bi=np.array([
        [p1[2,3]*x1-p1[0,3]],[p1[2,3]*y1-p1[1,3]],[p2[2,3]*x2-p2[0,3]],[p2[2,3]*y2-p2[1,3]]
    ])
    #求最小平方解，函數返回的第一項為我們要的
    result=np.linalg.lstsq(Ai,bi,rcond=None)[0]
    #把結果用x矩陣串起來
    X.append(result)
  #x 變成numpy  
  X=np.array(X)
  #todo6的第3點
  mask=np.array(X[:,2])>0
  #如果全部>0就是我們要的
  if mask.all()==True:
    bestp=p2
    x_best=np.array(X)
#出來的shape為(10，3，1)變成(10，3)    
x_best=x_best.reshape(x_best.shape[:2])
#為了要投影，在最後一個row加一排1    
Mi=np.vstack((np.transpose(x_best),np.ones(x_best.shape[0])))
#投影
mi=p1@Mi
mi_=bestp@Mi
mi=np.vstack((mi[0]/mi[2],mi[1]/mi[2]))
mi_=np.vstack((mi_[0]/mi_[2],mi_[1]/mi_[2]))
dis1=pts1-mi.T
dis2=pts2-mi_.T
#算出reconstruction error
error1=np.sum(np.sqrt(dis1[:,0]**2+dis1[:,1]**2))/pts1.shape[0]
error2=np.sum(np.sqrt(dis2[:,0]**2+dis2[:,1]**2))/pts2.shape[0]
print(f'reconstruction error1:{error1}\nreconstruction error2:{error2}')
```

```python
# TODO 7 (5 points): Visualize the 3D reconstruction.
#為了畫圖，找到每個點的最小值
x_min,x_max=np.min(x_best[:,0]),np.max(x_best[:,0])
y_min,y_max=np.min(x_best[:,1]),np.max(x_best[:,1])
z_min,z_max=np.min(x_best[:,2]),np.max(x_best[:,2])
fig=plt.figure(figsize=(15, 6))
#繪製左邊圖
ax=fig.add_subplot(1,2,1,projection='3d')
#畫出3d的頂點
ax.scatter3D(x_best[:,0],x_best[:,1],x_best[:,2],c='b',marker='o')
edges = [(0, 1), (1, 2), (2, 3), (3, 0), (4, 5), (5, 6), (0, 4), (1, 5), (2, 6)]
#根據邊緣的順序畫出在3d中的邊緣
for edge in edges:
  ax.plot(x_best[edge, 0], x_best[edge, 1], x_best[edge, 2], c='black')
#設定左邊圖仰角和方位角  
ax.view_init(elev=-90,azim=-115)

ax.set_title('Figure 5:\n ax.view_init(elev=-90,\n azim=-115)')
#繪製右邊圖
ax=fig.add_subplot(1,2,2,projection='3d')
#跟前面一樣
ax.scatter3D(x_best[:,0],x_best[:,1],x_best[:,2],c='b',marker='o')
edges = [(0, 1), (1, 2), (2, 3), (3, 0), (4, 5), (5, 6), (0, 4), (1, 5), (2, 6)]
for edge in edges:
  ax.plot(x_best[edge, 0], x_best[edge, 1], x_best[edge, 2], c='black')
#設定左邊圖仰角和方位角 
ax.view_init(elev=-139,azim=-113)
ax.set_title('Figure 6:\n ax.view_init(elev=-139,\n azim=-113)')

plt.show()
```
