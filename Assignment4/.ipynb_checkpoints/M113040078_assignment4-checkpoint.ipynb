```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

img1 = plt.imread('box_left_2k.jpg')
img2 = plt.imread('box_right_2k.jpg')

# 相機座標點
pts1 = np.array([[1148, 351],[1441, 481],[1427, 662],[1172, 525],[1524, 80],[1795, 134],[1744, 307],[1564, 320],[1579, 306],[1584, 288]])
pts2 = np.array([[747, 397],[944, 522],[1010, 704],[822, 558],[1212, 116],[1440, 170],[1433, 350],[1131, 360],[1152, 342],[1166, 326]])

# p1, p2
p1 = np.hstack((pts1, np.ones((pts1.shape[0], 1))))
p2 = np.hstack((pts2, np.ones((pts2.shape[0], 1))))

# 相機內參數
K1 = np.array([[1062.04, 0, 1098.52],[0, 1061.74, 599.665],[0, 0, 1]])
K2 = np.array([[1060.59, 0, 1127.03],[0, 1060.06, 645.711],[0, 0, 1]])

# TODO 1 (15 points): Normalize image points.

# mean of pts
t1 = np.mean(pts1, axis=0)
t2 = np.mean(pts2, axis=0)

# distance of pts to t
d1 = np.sqrt(np.sum((pts1 - t1)**2, axis=1))
d2 = np.sqrt(np.sum((pts2 - t2)**2, axis=1)) 

# scale factor
s1 = np.sqrt(2) / np.mean(d1)
s2 = np.sqrt(2) / np.mean(d2) 

# normalized pts
pts1_normal = s1 * (pts1 - t1)
pts2_normal = s2 * (pts2 - t2)  
                      
# T: translation matrix
T1 = np.array([[s1, 0, -s1 * t1[0]],
               [0, s1, -s1 * t1[1]],
               [0, 0, 1]])

T2 = np.array([[s2, 0, -s2 * t2[0]],
               [0, s2, -s2 * t2[1]],
               [0, 0, 1]])

# q = Tp
q1 = np.dot(T1, p1.T)
q2 = np.dot(T2, p2.T)

# origin
origin1 = np.mean(q1, axis=1)
origin2 = np.mean(q2, axis=1)

# mean distance
mean_distance1 = np.mean(np.sqrt(np.sum((q1.T - origin1)**2, axis=1)))
mean_distance2 = np.mean(np.sqrt(np.sum((q2.T - origin2)**2, axis=1)))

# display the mean distance of the image points in pts1_normal and pts2_normal
print('mean distance of the image points in pts1_normal: ', mean_distance1)
print('mean distance of the image points in pts2_normal: ', mean_distance2)

```

```python
# TODO 2 (25 points): Compute the fundamental matrix F.

A = np.zeros((pts1_normal.shape[0], 9))
for i in range(pts1_normal.shape[0]):
    A[i] = [pts1_normal[i][0] * pts2_normal[i][0], pts2_normal[i][0] * pts1_normal[i][1], pts2_normal[i][0],
        pts1_normal[i][0] * pts2_normal[i][1], pts1_normal[i][1] * pts2_normal[i][1], pts2_normal[i][1],
        pts1_normal[i][0], pts1_normal[i][1], 1]
   
# SVD
U,D,Vt = np.linalg.svd(A)
F = Vt[-1,:].reshape(3,3)
U,D,Vt = np.linalg.svd(F)
F_head = U @ np.array([[D[0],0,0],[0,D[1],0],[0,0,0]]) @ Vt

# Denormalize F (Fundamental Matrix)
F_head_de = T2.T @ F_head @ T1

# 計算 epipolar error:
error = 0
for i in range(pts1.shape[0]):
  x1,y1,_ = p1[i]
  x2,y2,_ = p2[i]
  error += np.array([x2,y2,1]) @ F_head_de @ np.array([x1,y1,1])
  
print("epipolar error:",error)  
```

```python
# TODO 3 (15 points): Draw epipolar lines in image1.

figure1 = img1.copy()
figure2 = img2.copy()
plt.figure()

# 1. 右邊
plt.subplot(1,2,2)

# 畫星號
plt.scatter(p2[0,0],p2[0,1],c='b',marker='*',s=10)
plt.scatter(p2[1,0],p2[1,1],c='orange',marker='*',s=10)
plt.scatter(p2[2,0],p2[2,1],c='g',marker='*',s=10)

# 去除刻度
plt.xticks([])
plt.yticks([])

# X軸標籤
plt.xlabel('Figure 2: right image')
plt.imshow(figure2)

# 2. 左邊
plt.subplot(1,2,1)

# e = F.T * p2
epipolarLine1 = F_head_de.T @ p2.T

# x
x_todo3=np.arange(img1.shape[1])

# y = (-a*x - c) / b
y1=(-epipolarLine1[0][0]*x_todo3 - epipolarLine1[2][0]) / epipolarLine1[1][0]
y2=(-epipolarLine1[0][1]*x_todo3 - epipolarLine1[2][1]) / epipolarLine1[1][1]
y3=(-epipolarLine1[0][2]*x_todo3 - epipolarLine1[2][2]) / epipolarLine1[1][2]

# line
plt.plot(x_todo3,y1,c='b')
plt.plot(x_todo3,y2,c='orange')
plt.plot(x_todo3,y3,c='g')

# 去除刻度
plt.xticks([])
plt.yticks([])

# X軸標籤
plt.xlabel('Figure 1: left image')
plt.imshow(figure1)

plt.show()
```

```python
# TODO 4 (10 points): Draw the epipole and epipolar lines in image1.

plt.figure()

# Fe = 0,we need to find the null space of F,but F rank is 2,so we use SVD to find the Vt's last row
U,E,Vt = np.linalg.svd(F_head_de)

# e is the last row of Vt,and we need to normalize it
e = np.array([int(Vt[-1,0] / Vt[-1,2]), int(Vt[-1,1] / Vt[-1,2])])

# 白色畫布
table = np.ones((img1.shape[0], e[0]+300, 3), np.uint8) * 255

# 填入img1
table[:,:img1.shape[1]] = img1

# x
x_todo4=np.arange(e[0])

# y = (-a*x - c) / b
y1=(-epipolarLine1[0][0]*x_todo4 - epipolarLine1[2][0]) / epipolarLine1[1][0]
y2=(-epipolarLine1[0][1]*x_todo4 - epipolarLine1[2][1]) / epipolarLine1[1][1]
y3=(-epipolarLine1[0][2]*x_todo4 - epipolarLine1[2][2]) / epipolarLine1[1][2]

# line
plt.plot(x_todo4,y1,c='b')
plt.plot(x_todo4,y2,c='orange')
plt.plot(x_todo4,y3,c='g')

# epipole
plt.scatter(e[0],e[1],c='r',marker='x')

# 去除刻度
plt.xticks([])
plt.yticks([])

# X軸標籤
plt.xlabel('Figure 3')
plt.imshow(table)

plt.show()
```

```python
# TODO 5 (5 points): Compute the 4 possible camera matrices.

E = K2.T @ F_head_de @ K1

# SVD
U,D,Vt = np.linalg.svd(E)

# Let u3 be the last column of U,
u3 = U[:,-1]

# W 
W = np.array([[0,-1,0],[1,0,0],[0,0,1]])

# If det(UWVt) < 0, then replace WW with −WW.
if np.linalg.det(U @ W @ Vt) < 0:
  W = -W

# P = [I|0]  
P = np.hstack((np.identity(3), np.zeros(shape=(3,1))))

# r1 r2 t1 t2
R1 = U @ W @ Vt
R2 = U @ W.T @ Vt
t1 = np.expand_dims(u3,axis=1)
t2 = np.expand_dims(-u3,axis=1)

# extrinsic matrices list
Rt2_list=[np.hstack((R1,t1)),np.hstack((R1,t2)),np.hstack((R2,t1)),np.hstack((R2,t2))]
```

```python
# TODO 6 (25 points): Triangulation.
#        There is no need to use for loops when computing the reconstruction errors of pts1 (or pts2).
#        If you do, you can get at most 20 point.

for j in range(4):
    # Let P1 = K1[I, 0], and P2 = K2[R,t], where [R,t] is Rt2_list[j].
    P1 = K1 @ P
    P2 = K2 @ Rt2_list[j]
    M = np.empty((0,3))
    for i in range(pts1.shape[0]):
        # Given P1, P2, pts1, pts2, implement the triangulation method in page 8 of 08 Two-View Geometry.pdf to obtain [Xi, Yi, Zi].
        Ai = np.array([[P1[0,0] - P1[2,0]*p1[i][0], P1[0,1] - P1[2,1]*p1[i][0], P1[0,2] - P1[2,2]*p1[i][0]],
                       [P1[1,0] - P1[2,0]*p1[i][1], P1[1,1] - P1[2,1]*p1[i][1], P1[1,2] - P1[2,2]*p1[i][1]],
                       [P2[0,0] - P2[2,0]*p2[i][0], P2[0,1] - P2[2,1]*p2[i][0], P2[0,2] - P2[2,2]*p2[i][0]],
                       [P2[1,0] - P2[2,0]*p2[i][1], P2[1,1] - P2[2,1]*p2[i][1], P2[1,2] - P2[2,2]*p2[i][1]]])
        
        bi = np.array([[P1[2,3]*p1[i][0] - P1[0,3]],
                       [P1[2,3]*p1[i][1] - P1[1,3]],
                       [P2[2,3]*p2[i][0] - P2[0,3]],
                       [P2[2,3]*p2[i][1] - P2[1,3]]])
        
        # np.linalg.lstsq()
        Xi, Yi, Zi = np.linalg.lstsq(Ai, bi, rcond=None)[0]
        M = np.vstack((M, np.array([Xi, Yi, Zi]).T)) # 10*3  
    
    # Find index j such that X_list[j][:,2] are all positive, and denote X_list[j] by X_best.
    if np.all(M[:,2] > 0):
        break
    
    
M_head = np.hstack((M, np.ones((M.shape[0],1)))) # 10*4
m1_head = P1 @ M_head.T 
m2_head = P2 @ M_head.T 
m1 = m1_head[:2, :] / m1_head[2, :]
m2 = m2_head[:2, :] / m2_head[2, :]

# reconstruction error
error1 = np.sum(np.sqrt(np.sum((pts1.T - m1)**2, axis=1))) / pts1.shape[0]
error2 = np.sum(np.sqrt(np.sum((pts2.T - m2)**2, axis=1))) / pts2.shape[0]
print("reconstruction error1:",error1)
print("reconstruction error2:",error2)
```

```python
# TODO 7 (5 points): Visualize the 3D reconstruction.
x_min, x_max = np.min(M[:,0]), np.max(M[:,0])
y_min, y_max = np.min(M[:,1]), np.max(M[:,1])
z_min, z_max = np.min(M[:,2]), np.max(M[:,2])

fig=plt.figure()

# 左圖
ax = fig.add_subplot(1,2,1,projection='3d')
# 畫點 & 邊緣
ax.scatter3D(M[:,0], M[:,1], M[:,2], c='b', marker='.')
edges = [(0, 1), (1, 2), (2, 3), (3, 0), (4, 5), (5, 6), (0, 4), (1, 5), (2, 6)]
for edge in edges:
  ax.plot(M[edge,0], M[edge,1], M[edge,2], c='black', linewidth=0.5)
# 仰角和方位角  
ax.view_init(elev=-90, azim=-115)
ax.set_xlim([x_min-0.1, x_max+0.1])
ax.set_ylim([y_min-0.1, y_max+0.1])
ax.set_zlim([z_min-0.1, z_max+0.1])
ax.set_title("Figure 5:\nax.view_init(elev=-90,\nazim=-115)", y=-0.4, fontsize=10)

# 右圖
ax=fig.add_subplot(1,2,2,projection='3d')
# 畫點 & 邊緣
ax.scatter3D(M[:,0],M[:,1],M[:,2],c='b',marker='.')
edges = [(0, 1), (1, 2), (2, 3), (3, 0), (4, 5), (5, 6), (0, 4), (1, 5), (2, 6)]
for edge in edges:
  ax.plot(M[edge, 0], M[edge, 1], M[edge, 2], c='black', linewidth=0.5)
# 仰角和方位角 
ax.view_init(elev=-139, azim=-113)
ax.set_xlim([x_min-0.1, x_max+0.1])
ax.set_ylim([y_min-0.1, y_max+0.1])
ax.set_zlim([z_min-0.1, z_max+0.1])
ax.set_title("Figure 6:\nax.view_init(elev=-139,\nazim=-113)", y=-0.4, fontsize=10)

plt.show()
```
