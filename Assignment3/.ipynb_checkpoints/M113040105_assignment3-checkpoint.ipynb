```python
import numpy as np
import cv2
import time
import matplotlib.pyplot as plt
%matplotlib inline

# TODO 1 (15%): Implement calc_homography()
def calc_homography(pts_src, pts_dst):
  # enter your code here, and remove the following line
    A=[]
    for i in range(pts_src.shape[0]):
        x,y=pts_src[i]
        u,v=pts_dst[i]
        A.append([-x,-y,-1,0,0,0,x*u,y*u,u])
        A.append([0,0,0,-x,-y,-1,x*v,y*v,v])    
    A=np.array(A,dtype=np.float64)
    A=A.T@A
    u, d, ut = np.linalg.svd(A)
    out=u[:,-1].reshape(3,3)
    return out
```

```python
img_src = cv2.imread("monet.jpg")
img_dst = cv2.imread("painting.jpg")
ht_src, wid_src, _ = img_src.shape
ht_dst, wid_dst, _ = img_dst.shape
pts_src = np.array([[0,0],[wid_src, 0],[wid_src,ht_src],[0,ht_src]])
pts_dst = np.array([[299,515],[1776,360],[1680,1834], [239,1615]])
print('source_shape:', ht_src, wid_src)
print('destination_shape:', ht_dst, wid_dst)

H = calc_homography(pts_src, pts_dst)
H = H/H[2,2]
print(H)
```

```python
# TODO 2 (10%): Project the source image on the frame
#         formed by v1, v2, v3, and v4 in the target image via direct mapping.
# Display the projection result.
projection_result=img_dst.copy()
for j in range(ht_src):
    for i in range(wid_src):
        points=np.array([i,j,1])
        x,y,z=H@points
        x,y= int(x/z), int(y/z)
        if y<ht_dst and x<wid_dst:
            projection_result[y,x]=img_src[j,i]          
plt.imshow(projection_result[:,:,::-1])
```

```python
# TODO 3 (20%): Project the source image on the frame
#         formed by v1, v2, v3, and v4 in the target image via inverse mapping.
# Display the projection result.
G=np.linalg.inv(H)
inverse_projection_result=img_dst.copy()
img_poly=np.zeros_like(img_dst)
img_poly=cv2.fillPoly(img_poly,[pts_dst],(255,255,255))
inverse_projection_result=img_dst.copy()
for j in range(ht_dst):
    for i in range(wid_dst):
        if img_poly[j,i,0]>0:
            points=np.array([i,j,1])
            x,y,z=G@points
            x,y= int(x/z), int(y/z)
            if y<ht_src and x<wid_src:
                inverse_projection_result[j,i]=img_src[y,x]          
plt.imshow(inverse_projection_result[:,:,::-1])
```

```python
# TODO 5 (20%): Warping library2.jpg to create img_warp.
# Display the warped result.
H = calc_homography(pts_dst, pts_src)
H = H/H[2,2]
img_canvas=np.zeros(shape=(ht_src, wid_dst+wid_src, 3),dtype=np.uint8)
img_warp=img_canvas.copy()
img_poly=np.zeros((ht_src, wid_dst+wid_src))
img_poly[:,wid_dst:] = 1
max_=0
for j in range(ht_src):
    for i in range(wid_dst+wid_src):
        if img_poly[j,i]>0:
            points=np.array([i,j,1])
            x,y,z=H@points
            x,y= int(x/z), int(y/z)
            if 0<=y<ht_src and  0<=x <wid_src:
                img_warp[j,i]=img_src[y,x]
                if i>max_:
                    max_=i           
plt.imshow(img_warp[:,:,::-1])
```

```python
# TODO 6 (5%): Stitching library1.jpg and img_warped.
# Display the stitched result.
img_stitched=img_warp.copy()
img_stitched[:, :wid_dst] = img_dst
plt.imshow(img_stitched[:,:,::-1])
plt.show()
```

```python
# TODO 7 (10%): Post-processing the stitched result.
# Display the final result.
img_stitched_final=img_stitched[:,:max_,:]
plt.imshow(img_stitched_final[:,:,::-1])
plt.show()
```

```python
# TODO 8 (10%): Complete the same task as in TODO 3, but without using any for loops.
img_src = cv2.imread("monet.jpg")
img_dst = cv2.imread("painting.jpg")
ht_src, wid_src, _ = img_src.shape
ht_dst, wid_dst, _ = img_dst.shape
pts_src = np.array([[0,0],[wid_src, 0],[wid_src,ht_src],[0,ht_src]])
pts_dst = np.array([[299,515],[1776,360],[1680,1834],[239,1615]])
# enter your code here
# compute the homography matrix
H= calc_homography(pts_dst, pts_src)
H=H/H[2,2]

img_poly=np.zeros_like(img_dst)
img_poly=cv2.fillPoly(img_poly,[pts_dst],(255,255,255))
roi_poly=np.argwhere(img_poly>0)

roi_dst=np.zeros((2,roi_poly.shape[0]))
roi_dst[0,:]=roi_poly[:,1]
roi_dst[1,:]=roi_poly[:,0]

roi_src=np.vstack((roi_dst,np.ones((1,roi_dst.shape[1]))))
roi_src=H @ roi_src
roi_src=roi_src[:2,:]/roi_src[2,:]

idx=(roi_src[0,:]<wid_src)&(roi_src[1,:]<ht_src)
img_out=img_dst.copy()
img_out[(roi_dst[1,idx]).astype(int),(roi_dst[0,idx]).astype(int)]=img_src[(roi_src[1,idx]).astype(int),(roi_src[0,idx]).astype(int)]

plt.imshow(img_out[:,:,::-1])
```
