```python
import numpy as np
import cv2
import time
import matplotlib.pyplot as plt
%matplotlib inline

# TODO 1 (15%): Implement calc_homography()
def calc_homography(points_source, points_destination):
  
  A = [] # 2n x 9 matrix
  
  # Iterate over all points
  for i in range(len(points_source)):
      
      # Extract x, y, u, v from points_source and points_destination
      x, y = points_source[i][0], points_source[i][1]
      u, v = points_destination[i][0], points_destination[i][1]
      
      # Append to A
      A.append([-x, -y, -1, 0, 0, 0, u*x, u*y, u])
      A.append([0, 0, 0, -x, -y, -1, v*x, v*y, v])
 
  # Convert A to numpy array
  A = np.asarray(A,dtype = np.float64)
  
  # Compute SVD of A.T@A
  U, D, Ut = np.linalg.svd(A.T@A)

  # Compute h
  h = Ut[-1]

  # Reshape h to 3x3
  H = h.reshape(3, 3)
  
  return H
```

```python
img_source = cv2.imread("monet.jpg")
img_destination = cv2.imread("painting.jpg")

# img_src is numpy arrays of shape (902, 1457, 3). monet.jpg
# img_dst is numpy arrays of shape (2163, 2089, 3). painting.jpg
height_source, width_source, _ = img_source.shape
height_destination, width_destination, _ = img_destination.shape

# pts_src and pts_dst are numpy arrays both of shape (4,2).
points_source = np.array([    
    [0,0],                          #top left
    [width_source, 0],              #top right
    [width_source,height_source],   #bottom right 
    [0,height_source]               #bottom left
    ])
points_destination = np.array([
    [299,515],                      #top left
    [1776,360],                     #top right
    [1680,1834],                    #bottom right
    [239,1615]                      #bottom left
    ])

print('source_shape:', height_source, width_source)
print('destination_shape:', height_destination, width_destination)

H = calc_homography(points_source, points_destination)
H = H / H[2,2]

print(H)
```

```python
# TODO 2 (10%): Project the source image on the frame formed by v1, v2, v3, and v4 in the target image via direct mapping.

# Create a copy of the destination image.
projection_result = img_destination.copy()

# Iterate over all pixels in the source image.
for j in range(height_source):
    for i in range(width_source):

        # Compute the corresponding pixel[x, y, z] in the source pixel[i, j, 1].
        x, y, z = H @ np.array([i, j, 1])

        # Normalize the coordinates.
        # z is the depth of the pixel in the source image.
        x, y = int(x/z), int(y/z)

        # Copy the pixel value from the source image to the destination image.
        if y < height_destination and x < width_destination:
            projection_result[y, x] = img_source[j, i]

# Display the projection result.
plt.imshow(cv2.cvtColor(projection_result, cv2.COLOR_BGR2RGB))
plt.show()

            
```

```python
# TODO 3 (20%): Project the source image on the frame formed by v1, v2, v3, and v4 in the target image via inverse mapping.

# Create a copy of the destination image.
projection_result = img_destination.copy()
H_inverse = np.linalg.inv(H)

# Iterate over all pixels in the destination image.
for j in range(height_destination):
    for i in range(width_destination):

        # Compute the corresponding pixel[x, y, z] in the destination pixel[i, j, 1].
        x, y, z = H_inverse @ np.array([i, j, 1])

        # Normalize the coordinates.
        # z is the depth of the pixel in the destination image.
        x, y = int(x/z), int(y/z)

        # Copy the pixel value from the source image to the destination image.
        if y < height_source and x < width_source and y >= 0 and x >= 0:
            projection_result[j, i] = img_source[y, x]

# Display the projection result.
plt.imshow(cv2.cvtColor(projection_result, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# TODO 5 (20%): Warping library2.jpg to create img_warp.

# Step 5.1 
# library2  library1  homography matrix
H = calc_homography(points_library2, points_library1)
H = H / H[2,2]

# Step 5.2
#  library1  library2  library1 
img_canvas = np.zeros((img_library1.shape[0], img_library1.shape[1] + img_library2.shape[1], 3), dtype=np.uint8)
height_source, width_source = img_library2.shape[:2]
height_destination, width_destination = img_canvas.shape[:2]

# Step 5.3 
# Similar to the steps in TODO 3, use H to project the whole library2.jpg on img_canvas[:,wid_dst:,:],and the result is denoted by img_warp as shown below.
img_warp = img_canvas.copy()
H_inverse = np.linalg.inv(H)

# Iterate over all pixels in the destination image.
for j in range(height_destination):
    for i in range(width_destination):

        # Compute the corresponding pixel[x, y, z] in the destination pixel[i, j, 1].
        x, y, z = H_inverse @ np.array([i, j, 1])

        # Normalize the coordinates.
        # z is the depth of the pixel in the destination image.
        x, y = int(x/z), int(y/z)

        # Copy the pixel value from the source image to the destination image.
        if y < height_source and x < width_source and y >= 0 and x >= 0:
            img_warp[j, i] = img_library2[y, x]

# Display the projection result.
plt.imshow(cv2.cvtColor(img_warp, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# TODO 6 (5%): Stitching library1.jpg and img_warped.
# Replace the first few columns of img_warped with library1.jpg to form the array img_stitched as shown in the following figure.
# Display the stitched result.
img_stitched = img_canvas.copy()
img_stitched[:, :600] = img_destination
img_stitched[:, 600:] = img_warp[:, 600:]
plt.imshow(cv2.cvtColor(img_stitched, cv2.COLOR_BGR2RGB))
plt.show()
```
